# tuple

```python
t = (2 , 4 , 5)
print(f'type = {type(t)}')
print(f'length = {len(t)}')

print('\n#####################\n')

#slice

t = (2 , 4 , 6 , 8 , 10 , 1 , 3)
print(t[0])
print(t[:4])
print(t[3 : 6])
print(t[4 :])
print(t[::-1])
print(t[ : 5 : 2])

print('\n#####################\n')

#index
t = ('amir' , 'gholam' , 'reza')
a = t.index('amir')
print(f'index of amir : {a}')
print('amir' in t)
print('gholam' not in t)

print('\n#####################\n')

#####
a = (3)
print(f'type of a : {type(a)}')
t = (3,)
print(f'type of t : {type(t)}')

print('\n#####################\n')

#max , min , sum , count
t = (1 , 2 , 3 , 4 , 5 , 6 , 1 , 1 , 1 , 2 , 2 , 2)
print(f'length a t : {t}')
print(f'sum of tuple argumant : {sum(t)}')
print(f'max = {max(t)}')
print(f'min = {min(t)}')
print(f'number of 1 : {t.count(1)}')

print('\n#####################\n')

#concat and reverse
t = (1 , 2 , 3)
a = t * 2
print(a)

b = tuple(reversed(t))
print(f'revrese : {b}')

print('\n#####################\n')

#Comparison of tuples
a = (1 , 2)
b = (2 , 1)
c = (1 , 2)

print(f'comparison of a and b : {a == b}')
print(f'comparison of a and c : {a == c}')
```
