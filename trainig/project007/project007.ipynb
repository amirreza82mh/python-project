```python
#slice
a = [1 , 4 , 4 , 23 , 55 , 23 , 14 , 23]
print(a[:])
print(a[0 : 3])
print(a[3 : 8])
print(a[6 : 3])
print(a[ :: -1])
#step slice
print('\n############################\n')
print(a[3 : 9 : 2])
print(a[: 7 : 3])
print(a[4 :: 1])
print(a[:: 2])

#concat
a = [1 , 2 , 3]
b = a * 2
print(b)

a = [1 , 2 , 3]
b = [4 , 5]
c = a + b
print(c)

print('\n############################\n')

#list in the list
a = [1 , 2 , 3 ,4]
print(3 in a)
print(5 not in a)

a = [1 , 2 , [3 , 4] , 5]
print(a)
print(a[2])
print(a[2][0])
```

```python
# min max
a = [2 , 12 , 13 , 35 , 30 , 1]
m = a[0]
for i in a:
  if i > m:
    m = i
print(f'max = {m}')

m = a[0]
for i in a:
  if i < m:
    m = i
print('min = {:d}'.format(m))
```

```python
# list function

#insert and remove
a = [1 , 2 , 3 , 4 , 4 , 5]
print('count {:d}'.format(a.count(4)))
a.insert(4 , 5)
a.insert(5 , 6)
print(a)
a.remove(4)
a.remove(4)
a.remove(5)
a.remove(5)
print(a)
a.insert(3 , 4)
a.insert(4 , 5)
print(a)

print('\n############################\n')

#pop
a = [1 , 2 , 3 , 4 , 5]
x = a.pop()
print(a)
print('pop = {:d}'.format(x))
print('len = {:d}'.format(len(a)))
x = a.pop()
print(a)
print('pop = {:d}'.format(x))
print('len = {:d}'.format(len(a)))

print('\n############################\n')

#delet
a = [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8]
del a[1]
print(a)
del a[1 : 5]
print(a)

print('\n############################\n')

#reverse 
a = [1 , 2 , 3 , 4]
print(a)
a.reverse()
print(a)
a.reverse()
print(a)
print(a[:: -1])

print('\n############################\n')

#sort
a = [2 , 10 , 3 , 3 , 6 , 5 , 12 , 11]
a.sort()
print(a)
a.reverse()
print(a)

print('\n############################\n')

#extend
a = [1 , 2 , 3]
b = [4 , 5]
a.extend(b)
print(a)
print('len = {:d}'.format(len(a)))
print(b)
print(f'len = {len(b)}')

print()

a = [1 , 2 , 3]
b = [4 , 5]
b.extend(a)
print(a)
print('len = {:d}'.format(len(a)))
print(b)
print(f'len = {len(b)}')

print('\n############################\n')

#append
a = [1 , 2 , 3]
b = [4 , 5]
a.append(b)
print(a)
print(f'len = {len(a)}')

print()

a = [1 , 2 ,3]
b = [4 , 5]
b.append(a)
print(b)
print(f'len = {len(b)}')

print('\n############################\n')

#clear
a = [1 , 2 , 3 , 4 , 5]
print(f'befor clear : {a}')
a.clear()
print(f'after clear : {a}')

print('\n############################\n')

#copy
a = [1 , 2 , 3 , 4]
b = a.copy()
print(f'a = {a}')
print(f'b = {b}')
```

```python
# ----matrix----

a = [
    [1 , 2 , 3],
    [4 , 5 , 6],
    [7 , 8 , 9] 
     ]
print(a)
print(f'length = {len(a)}')
print(a[0])
print(a[2])

print()

for i in a:
  print(i)

for i in range(3):
  print(a[i][i] , end = ' ')

print()

for i in range(3):
  print(a[1][i] , end = ' ')

print()

for i in range(3):
  print(a[i][2-i] , end = ' ')

print()

print('\n############################\n')

b = []
b.extend([sum(i) for i in a])
print(b , end = ' ')

print()

c = []
for col in range(3):
    c.append(sum(i[col] for i in a))
print(c , end = ' ')    
print()

print('\n############################\n')

x = 2
y = x
y += 1
print(x)   # 2
print(y)   # 3

x = []
y = x
y.append(5)
print(x)    # [5]
print(y)    # [5]
```
